```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

# Exemple d'utilisation
pdf_path = r"C:\Users\Utilisateur\Mon_Premier_Projet_Python.pdf"
text = extract_text_from_pdf(pdf_path)
print(text)

def count_words(text):
    words = text.split()
    return len(words)

word_count = count_words(text)
print(f"Nombre de mots : {word_count}")

import re
print(len(re.split(r",;.| ", text)))
res = {}
for i in text.split():
  if i not in res:
    res[i] = 1
  else:
    res[i] += 1

{mot: text.split().count(mot) for mot in set(text.split())}

#dict(sorted({mot: texte.split().count(mot) for mot in set(texte.split())}.items(), key=lambda x: x[1], reverse=True))
```
